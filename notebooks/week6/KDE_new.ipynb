```python
nsample = 2**11
nmc = 2**21
numruns = int(nmc/nsample)

x_h0_all = []
x_h1_all = []

for i in tqdm(range(numruns)):
    x_h0_all.append(simulator.sample(nsample)['x0']-simulator.sample(nsample)['mu'])
    x_h1_all.append(simulator.sample(nsample)['xi']-simulator.sample(nsample)['mu'])

x_h0_all = torch.concatenate(x_h0_all).flatten()
x_h1_all = torch.concatenate(x_h1_all).flatten()
```

```python
kdebloc = np.load('../../data_bin/KDE_ref/KDE_archive.npz')
dat_h0 = kdebloc['dat_h0']+1e-21
dat_h1 = kdebloc['dat_h1']+1e-21
```

```python
fig, ax1 = pf.create_plot(size=(4,2))
ax2 = fig.add_axes((0,1,1,1), sharex=ax1)
plt.setp(ax2.get_xticklabels(), visible=False)

grid = np.linspace(-30,30, 1000)
ti = -2 * (np.log(dat_h0)-np.log(dat_h1))
t0 = -2 * (np.log(dat_h0)-0)

ax1.plot(grid,dat_h0, color='#ff004f', label=r'H$_0$', lw=3)
ax1.plot(grid,dat_h1, color='#77aca2', label=r'H$_i$', lw=3)
ax1.legend()
ax1.set_xlim([-7,7])
ax1.set_ylabel(r'$p(x|...)$')

ax2.plot(grid,ti, lw=2,color='black')
ax2.set_ylabel(r'$-2\:\log\frac{p(x|H_i)}{p(x|H_0)}$')

pf.fix_plot([ax1,ax2])
plt.tight_layout()
plt.savefig(f'figs/newkde.png', dpi=700, bbox_inches = 'tight')
```

```python
from scipy.interpolate import interp1d
from scipy.stats import chi2

N_mc = 5000000
t0_erpolator = interp1d(grid, t0)
ti_erpolator = interp1d(grid, ti)
million = np.linspace(-2,6,N_mc)

ti_big  = ti_erpolator(million)
t0_big = t0_erpolator(million)

fig, ax = pf.create_plot()
ax2 = fig.add_axes((1.2,0,1,1), sharex=ax)

ax2.hist(ti_big-min(ti_big), bins=1000, density=True, color='#ff004f')
ax.hist(t0_big-min(t0_big), bins=1000, density=True, color='#77aca2')

ax.set_ylim([0,1])
ax2.set_ylim([0,.8])
ax.set_xlim([-.5,10])


ax.set_xlabel(r'$t_i=-2\:\log\frac{p(x|H_0)}{p(x|H_i)}\; [H_i]$')
ax2.set_xlabel(r'$t_i=-2\:\log\frac{p(x|H_0)}{p(x|H_i)}\; [H_0]$')
ax.set_ylabel(r'Density')

pf.fix_plot([ax,ax2])
```
